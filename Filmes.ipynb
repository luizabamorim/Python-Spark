Gênero mais frequente


Split para separar os gêneros

```python
dados_titulos.withColumn('gen_sep', f.split(f.col('genres'), ',')).limit(5).toPandas()

```

```python
dados_titulos.withColumn('gen_sep',f.split(f.col('genres'),',')).withColumn('generos',f.explode('gen_sep')).limit(5).toPandas()
```

Separação dos gêneros dos Filmes (um gênero por coluna)

```python
dados2=(dados_titulos.withColumn('gen_sep',f.split(f.col('genres'),',')).withColumn('generos',f.explode('gen_sep')))
```

```python
dados2.limit(5).toPandas()
```

```python
dados2.groupby('generos').agg(f.count('generos').alias('Contagem'),).orderBy(f.col('Contagem').desc()).toPandas()
```
